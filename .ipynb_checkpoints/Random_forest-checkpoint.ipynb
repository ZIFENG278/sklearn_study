```python
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
print(wine.data)
print(wine.target)
print(wine.data.shape, wine.target.shape)
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data
                                                ,wine.target
                                                ,test_size=0.3
                                                ,random_state=30
                                               )

```

```python
print(Xtrain.shape)
print(Ytrain.shape)
```

```python
clf = DecisionTreeClassifier(random_state=0)
rfc = RandomForestClassifier(random_state=0)


clf = clf.fit(Xtrain, Ytrain)
rfc = rfc.fit(Xtrain, Ytrain)

score_c = clf.score(Xtest, Ytest)
score_r = rfc.score(Xtest, Ytest)
```

```python
print(score_c, score_r)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score

clf_1 = []
rfc_1 = []

for i in range(10):
    rfc = RandomForestClassifier(n_estimators=25)
    rfc_s = cross_val_score(rfc,wine.data,wine.target,cv=10).mean() #交叉验证取平均
    rfc_1.append(rfc_s)
    clf = DecisionTreeClassifier()
    clf_s = cross_val_score(clf,wine.data,wine.target,cv=10).mean()
    clf_1.append(clf_s)
    
        
plt.plot(range(1,11),rfc_1,label = "Random Forest")
plt.plot(range(1,11),clf_1,label = "Decision Tree")
plt.legend()
plt.show()
```

```python
superpa = []

for i in range(200):
    rfc = RandomForestClassifier(n_estimators=i+1, n_jobs=-1)
    rfc_s = cross_val_score(rfc, wine.data, wine.target, cv=10).mean()
    superpa.append(rfc_s)
    #print(i)
```

```python
# print(max(superpa), superpa.index(max(superpa)))
plt.figure(figsize=[20,5])
plt.plot(range(1,11),superpa)
plt.show()
```
