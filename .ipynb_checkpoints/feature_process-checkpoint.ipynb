```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
```

```python
data_df = pd.DataFrame(data)
```

```python
data_df
```

```python
#归一化
scaler = MinMaxScaler()
scaler = scaler.fit(data)
result = scaler.transform(data)
```

```python
result
```

```python
result_ = scaler.fit_transform(data)  #功能和上面三行一样
```

```python
result_
```

```python
scaler.inverse_transform(result_)  #归一化后转回原来的数据
```

```python
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
```

```python
scaler = MinMaxScaler(feature_range=[5,10]) # 如果数据量太大就使用scaler = partial_fir(data)
result = scaler.fit_transform(data) 
result
```

```python
# 用numpy实现归一化也可以， bonus 
import numpy as np

x = np.array([[-1, 2], [-0.5, 6], [0, 10], [1, 18]])
```

```python
x_nor = (x - x.min(axis=0)) / (x.max(axis=0) - x.min(axis=0))
x_nor
```

```python
x_returned = x_nor * (x.max(axis=0) - x.min(axis=0)) + x.min(axis=0)
x_returned
```

```python
# 数据标准化standardization, Z-score normalization   x* = (x-u) / omiga  均值为0， 方差为1
from sklearn.preprocessing import StandardScaler
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
```

```python
scaler = StandardScaler()
scaler = scaler.fit(data)
print(scaler.mean_, scaler.var_)
```

```python
x_std = scaler.transform(data)
print(x_std.mean(), x_std.std())
```

```python
result = scaler.fit_transform(data)  # 实现上面三行的功能
result
```

```python
result = scaler.inverse_transform(result)
```

```python
result
```

```python
# 填补年龄
Age = data.loc[:, "Age"].values.reshape(-1,1) # -1意味着最后一行的数
```

```python
Age[:20]
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer() # 默认均值填充
imp_median = SimpleImputer(strategy="median") # 用中位数填充
imp_0 = SimpleImputer(strategy='constant', fill_value=0) # 用0填充
imp_mean = imp_mean.fit_transform(Age)
imp_median = imp_median.fit_transform(Age)
imp_0 = imp_0.fit_transform(Age)
```

```python
imp_mean[:20]
```

```python
imp_median[:20]
```

```python
imp_0[:20]
```

```python
# 使用中位数来填充Age
data.loc[:, "Age"] = imp_median
data.info()
```

```python
Embarked = data.loc[:, "Embarked"].values.reshape(-1, 1)
```

```python
imp_mode = SimpleImputer(strategy="most_frequent") # 填补众数
data.loc[:, "Embarked"] = imp_mode.fit_transform(Embarked)
data.info()
```

```python
data.head()
```

```python
# 把文字转变成数值型 标签专用
from sklearn.preprocessing import LabelEncoder

y = data.iloc[:, -1]

le = LabelEncoder()
label = le.fit_transform(y)
label
```

```python
le.classes_
```

```python
data.iloc[:, -1] = label
data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
#接口categories_对应LabelEncoder的接口classes_，一模一样的功能
data_ = data.copy()
data_.head()
```

```python
OrdinalEncoder().fit(data_.iloc[:,1:-1]).categories_  # 返回列中所有的特征， 特征专用
```

```python
data_.iloc[:,1:-1] = OrdinalEncoder().fit_transform(data_.iloc[:,1:-1])
data_.head()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
X = data.iloc[:,0].values.reshape(-1,1)
est = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
est.fit_transform(X)[:20]
```
