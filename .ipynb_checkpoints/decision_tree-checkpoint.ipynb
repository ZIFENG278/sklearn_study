```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size=0.3) # 30测试集
```

```python
Xtrain.shape
```

```python
wine.data.shape
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
```

```python
score
```

```python
import graphviz
```

```python
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度',
                '色调','od280/od315稀释葡萄酒','脯氨酸']

class_name = ['琴酒', '雪莉', '贝尔摩德']
```

```python

dot_data = tree.export_graphviz(clf
                                ,feature_names=feature_name
                                ,class_names=class_name
                                ,filled=True
                                ,rounded=True)

graph = graphviz.Source(dot_data)
graph
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                 ,random_state=30
                                 ,splitter="random"
                                 )
```

```python
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```

```python
import graphviz
dot_data = tree.export_graphviz(clf
                               ,feature_names= feature_name
                               ,class_names=["琴酒","雪莉","贝尔摩德"]
                               ,filled=True
                               ,rounded=True
                               )  
graph = graphviz.Source(dot_data)
graph
```

```python
import matplotlib.pyplot as plt  # 查看超参数

test = []

for i in range(10):
    clf = tree.DecisionTreeClassifier(max_depth=i+1
                                     ,criterion="entropy"
                                     ,random_state=30)
    clf = clf.fit(Xtrain, Ytrain)
    score = clf.score(Xtest, Ytest)
    test.append(score)
    
plt.plot(range(1,11), test, color="red", label="max_depth")
plt.legend()
plt.show()
```
